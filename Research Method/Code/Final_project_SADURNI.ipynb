```python
# Compute the YTM of a bond using the trading price, the par value, the number of years until maturity, the coupon rate,
# and the number of paiement by year. 

def YTM(trading_price, par_val, years_to_maturity, coupon_rate, frequency=1):
    
    # the total number of periods 
    nb_periods = years_to_maturity*frequency
    # the coupon paid each period
    coupon = coupon_rate*par_val/frequency
    dt = list(range(1,nb_periods+1))
    #the ytm computation with x unknown
    ytm = lambda x: - trading_price + (sum([coupon/(1+x/frequency)**(frequency*t) for t in dt]) + par_val/(1+x/frequency)**(frequency*years_to_maturity) )

    # optimization of ytm using the Newton algorithm
    return sco.newton(ytm,0.1)

print('YTM: ' + str(round(YTM(1069.69,1000,5,0.04)*100,3)) + ' %' )
rf=YTM(1069.69,1000,5,0.04)
```

##### 2. Using the 10 stocks from before, plot the efficient frontier (no short selling) and draw the capital market line (line connecting risk-free rate and the maximum Sharpe ratio portfolio).
```python
data = yf.download('ROP WFC TXT M SJM CMG DD CINF DVA MMC', start = '2010-01-01', end = '2020-01-01', interval = "1mo")
data = data['Adj Close'].dropna()

rets = np.log(data / data.shift(1))
rets
```

```python
def portfolio_ret(weights):
    return np.dot(rets.mean(), weights) * 12

def portfolio_vol(weights):
    return np.sqrt(np.dot(weights.T, np.dot(rets.cov(), weights)) * 12)

portfolio_rets = []
portfolio_vols = []

for p in range (10000):
    np.random.seed(p + 1)
    weights = np.random.rand(rets.shape[1])
    weights /= np.sum(weights)
    
    portfolio_rets.append(portfolio_ret(weights))
    portfolio_vols.append(portfolio_vol(weights))
    
portfolio_rets = np.array(portfolio_rets)
portfolio_vols = np.array(portfolio_vols)
```

```python
def min_sharpe(weights):
    return -portfolio_ret(weights) / portfolio_vol(weights)

cons = ({'type': 'eq',
         'fun': lambda x: np.sum(x) - 1})

bnds = tuple((0, 1) for x in range(rets.shape[1]))

equal_w = np.array(rets.shape[1] * [1 / rets.shape[1]])
print('equal weight: ' + str(equal_w[1]))
print('min_sharpe: ' + str(min_sharpe(equal_w)))
print('')

optimal = sco.minimize(min_sharpe,
                       equal_w,
                       method='SLSQP',
                       bounds=bnds,
                       constraints=cons)
optivol = sco.minimize(portfolio_vol,
                       equal_w,
                       method='SLSQP',
                       bounds=bnds,
                       constraints=cons)
print('optimal ;')
print(optimal)
print('')
print('optivol ;')
print(optivol)
```

```python
print('Optimal composition of the portfolio for maximum sharpe ratio : ' + str(optimal['x'].round(3)))
print('return : ' + str(portfolio_ret(optimal['x']).round(3)))
print('volatility : ' + str(portfolio_vol(optimal['x']).round(3)))
print('optimal_ratio : ' + str(portfolio_ret(optimal['x']) / portfolio_vol(optimal['x'])))
print('Optimal composition of the portfolio for minimum volatility : ' + str(optivol['x'].round(3)))
print('return : ' + str(portfolio_ret(optivol['x']).round(3)))
print('volatility : ' +str(portfolio_vol(optivol['x']).round(3)))
print('optivol_ratio : ' + str(portfolio_ret(optivol['x']) / portfolio_vol(optivol['x'])))
```

```python
cons = ({'type': 'eq',
         'fun': lambda x: portfolio_ret(x) - target_ret},
        {'type': 'eq',
         'fun': lambda x: np.sum(x) - 1})

bnds = tuple((0, 1) for x in weights)
target_rets = np.linspace(0.01, 0.2,50)
target_vols = []

for target_ret in target_rets:
    res = sco.minimize(portfolio_vol,
                       equal_w,
                       method='SLSQP',
                       bounds=bnds,
                       constraints=cons)
    target_vols.append(res['fun'])
```

```python
index = np.argmin(target_vols)
t_vols = target_vols[index:]
t_rets = target_rets[index:]

tck = sci.splrep(t_vols, t_rets)

def f(x):
    return sci.splev(x, tck, der=0)
         
def df(x):
    return sci.splev(x, tck, der=1)

def equations(p, rf=rf):
    equation_1 = rf - p[0]
    equation_2 = rf + p[1] * p[2] - f(p[2])
    equation_3 = p[1] - df(p[2])
    return equation_1, equation_2, equation_3

opt = sco.fsolve(equations, [rf, 2, 0.13])  
```

```python
max_weight = np.amax(result['x'].round(3))
max_weight_index = np.where(result['x'].round(3)==max_weight)
max_weight_stock = labels[int(max_weight_index[0])]

stock = max_weight_stock
index = np.where(np.array(labels)==max_weight_stock)
weights_stocks =np.array(result['x'].round(3))
weight = float(weights_stocks[index])

print('The maximum weighted stock of the portfolio is ' + max_weight_stock)
print('The risk free rate is ' + str(rf))
data_stock = rets[[stock]]
std_stock = data_stock.std()
std_stock = std_stock.values[0]
print('Standard deviation of the stock is ' + str(std_stock))
```

```python
###########################################################################################
###################################"PARAMETERS"############################################
###########################################################################################


stock_prices=data[max_weight_stock].to_numpy()
#stock prices at the end of the data set
S0 = stock_prices[-1]
print('Last stock price ' + str(round(S0,2)) + '$')

#the percentage of from which we want to hedge our position (for long put -> use a negative number, for long call -> use a positive number)
#here we want to long a put option so we use a negative number : 5%
beta=-5/100

#The strike price (beta% below S0)
K=S0*(1+beta)
print('Strike price : ' + str(round(K,2)) + '$')

#Date until maturity (in years)
T=0.5

#premium paid per option
premium = round(S0-S0*(1-0.5/100),2) #0.5% of the trading price of the stock (I use that because I dont know how premiums are computed)
print('Premium per contract : ' + str(premium) + '$')
```

```python
###########################################################################################
###################################"FUNCTIONS"#############################################
###########################################################################################

def long_call(S, K, premium):
    # Long Call Payoff = max(Stock Price - Strike Price, 0)
    P = list(map(lambda x: max(x - K, 0) - premium, S))
    return P

def long_put(S, K, premium):
    # Long Put Payoff = max(Strike Price - Stock Price, 0)    
    P = list(map(lambda x: max(K - x,0) - premium, S))
    return P

def long_payoff(S, K):
    P = list(map(lambda x: (x - K), S))
    return P 
```

```python
#let's assume that the price of the stock can go from S0/2 to S0*1.3 (of course it can goes below or above)
stock_prices = np.linspace(S0/2,S0*1.3,1000)

###########################################################################################
#####################################"OPTION"##############################################
###########################################################################################

payoffs = long_payoff(stock_prices,K)
#Change to long_call if you want to hedge with a call
puts = long_put(stock_prices,K,premium)

###########################################################################################
#####################################"PLOTs"###############################################
###########################################################################################

fig, (ax1, ax2, ax3) = plt.subplots(3, 1,
                               sharex=False,
                               figsize=(15, 15))

ax1.plot(stock_prices, payoffs, 'b')
ax1.axvline(K,c='r',lw=1)
ax1.legend(['Long payoff', 'K'])
ax1.grid()
ax1.axhline(c='k',lw=1)

ax1.set_xlabel('Stock price')
ax1.set_ylabel('Option payoff')
ax1.title.set_text("Payoff of a long put option ")

ax2.plot(stock_prices, puts, 'b')
ax2.axvline(K,c='r',lw=1)
ax2.legend(['Long put', 'K'])
ax2.set_xlabel('Stock price')
ax2.set_ylabel('Option payoff')
ax2.grid()

res=np.add(payoffs,puts)
ax3.plot(stock_prices, res, 'b')
ax3.grid()
ax3.set_xlabel('Stock price')
ax3.set_ylabel('Option payoff')
ax3.axvline(K,c='r',lw=1)
ax3.legend(['Hedge', 'K'])
ax3.title.set_text('Hedged position')
```

```python
###########################################################################################
###################################"BINOMIAL"##############################################
###########################################################################################

# Binomial formulas for call and put prices (seen in class but I have changed the parameters) 
def binomial(S0, K, std_stock, option_1='call', option_2='euro'):
    M = 100
    dt = T / M
    u = np.exp(std_stock * np.sqrt(dt))
    d = 1 / u
    bt = [[0 for j in range(i + 1)] for i in range(M + 1)]
    R = np.exp(rf * dt)
    q = (R - d) / (u - d)
    
    for i in range(M, -1, -1):
        for j in range(i + 1):
            if option_1 =='call':
                bt[i][j] = max(S0 * u**j * d**(i - j) - K, 0)
            else:
                bt[i][j] = max(-S0 * u**j * d**(i - j) + K, 0)

    for i in range(M-1, -1, -1):
        for j in range(i + 1):
            if option_2 =='euro':
                bt[i][j] = np.exp(-rf * dt) * (q * bt[i + 1][j + 1] + (1 - q) * bt[i + 1][j])
            else:
                bt[i][j] = max(np.exp(-rf * dt) * (q * bt[i + 1][j + 1] + (1 - q) * bt[i + 1][j]), bt[i][j])
                
    return bt[0][0]
```

```python
# Array with the payoffs using the binomial formula
payoff_binomial=[]
for i in stock_prices:
    payoff_binomial.append(binomial(i,K,std_stock,'put','euro'))
```

```python
###########################################################################################
####################################"PAYOFFs"##############################################
###########################################################################################

fig, (ax1,ax2) = plt.subplots(1,2, figsize=(10,5))

ax2.plot(stock_prices, payoff_binomial,'b')
ax2.set_xlabel('Stock Price')
ax2.title.set_text('Option payoff using binomial')
ax2.grid(True)
ax2.axvline(K)
ax2.legend(['Price', 'K'])
```

```python
alpha = weight*100                    #the percentage of the stock in the portfolio
delta = -30/100                           #the stock price moves of 30%
amount = max_weight*100000                #the amount of money corresponding to the stock in the portfolio
nb_shares = amount/S0                     #the number of shares in of the stock
premium_paid = round(nb_shares*premium,2) #the premium paid
stock_at_T=round(S0*(1+delta),2)           #the price of the stock at maturity
amount_at_T = nb_shares*stock_at_T        #the amount that we would have without the hedge
diff_amount = amount_at_T-amount          #the amount that we have would have won/lost without the hedge
diff = round(nb_shares*K-amount-premium_paid,2)   #the amount we lost/won
hedge_amount = round(diff-diff_amount,2)          #the amount that we hedged
print('The pourcentage of the stock in our portfolio is ' + str(alpha) + '%')
print('Which means ' + str(int(amount))+ str('$ of stocks'))
print('Which means ' + str(int(nb_shares)) + ' shares of ' + max_weight_stock + ' at a trading price of ' + str(round(S0,2)) + '$')
print('If the stock price goes to ' + str(stock_at_T) + '$, then we would have ' + str(round(amount_at_T,2)) + '$')
print('Which means that we would have won/lost ' + str(round(diff_amount,2)) + '$')
print('However we hedged our the position and needed to pay the premium for a price of ' + str(premium_paid) + '$')
print('We hedged our position at a strike price of ' + str(round(K,2)) +'$ , hence we *only* have won/lost ' + str(diff) + '$')
print('Which means that we hedged ' + str(hedge_amount) + '$')
```
